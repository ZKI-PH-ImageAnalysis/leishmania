## Camera Test
This notebook contains the following tests:
- Which camera ports are available
- Default, minimum and maximum values for config.yml
- Available resolutions

#### Check Available Ports

```python
# Number of Ports that are checked for opencv
num_ports = 10
```

```python
import cv2

available_ports = []
for port in range(num_ports):
    cap = cv2.VideoCapture(port)
    if cap.isOpened():
        available_ports.append(port)
        cap.release()

if available_ports:
    print("Available camera ports:", available_ports)
else:
    print("No cameras found on any of the checked ports.")
```

#### Property Test
- Returns current properties as default
- returns min and max values of the properties

```python
# Specify the port for the camera you intend to use.
used_port = 0
```

```python
import cv2

cap = cv2.VideoCapture(used_port)

default_list = {
        'Brightness': cap.get(cv2.CAP_PROP_BRIGHTNESS),
        'Contrast': cap.get(cv2.CAP_PROP_CONTRAST),
        'Saturation': cap.get(cv2.CAP_PROP_SATURATION),
        'Hue': cap.get(cv2.CAP_PROP_HUE),
    }

properties_list = {
        'Brightness': [],
        'Contrast': [],
        'Saturation': [],
        'Hue': [],
    }

properties = [cv2.CAP_PROP_BRIGHTNESS, cv2.CAP_PROP_CONTRAST, cv2.CAP_PROP_SATURATION, cv2.CAP_PROP_HUE]
if not cap.isOpened():
    print("Error opening the camera")
else:
    for value in range(-1000,1000):
        for property in properties:
            cap.set(property, value)
            actual_property = cap.get(property)
            if int(actual_property) == value:
                if properties[0] == property:
                    properties_list['Brightness'].append(value)
                elif properties[1] == property:
                    properties_list['Contrast'].append(value)
                elif properties[2] == property:
                    properties_list['Saturation'].append(value)
                elif properties[3] == property:
                    properties_list['Hue'].append(value)

print("Property: min - max")
for i in range(len(properties_list)):
    print(f'{list(properties_list.keys())[i]}: {min(properties_list[list(properties_list.keys())[i]])} - {max(properties_list[list(properties_list.keys())[i]])}')

print("\nProperty: default")
for i in range(len(default_list)):
    print(f'{list(default_list.keys())[i]}: {default_list[list(default_list.keys())[i]]}')
    cap.set(properties[i], list(default_list.values())[i])

cap.release()
```

#### Resolution Test (slow)

```python
import cv2

cap = cv2.VideoCapture(used_port)

if not cap.isOpened():
    print("Error opening the camera")
else:
    # List of common resolutions to test
    common_resolutions = [
        (320, 200),
        (320, 240),
        (340, 256),
        (480, 320),
        (640, 480),
        (648, 486),
        (680, 512),
        (720, 480),
        (720, 576),
        (800, 480),
        (800, 600),
        (854, 480),
        (1024, 600),
        (1024, 768),
        (1136, 768),
        (1280, 720),
        (1280, 800),
        (1280, 960),
        (1280, 1024),
        (1296, 972),
        (1360, 1024),
        (1400, 1050),
        (1440, 900),
        (1440, 960),
        (1440, 1080),
        (1600, 1200),
        (1680, 1050),
        (1920, 1080),
        (1920, 1200),
        (2048, 1080),
        (2048, 1536),
        (2560, 1440),
        (2560, 2048),
        (2592, 1944),
        (3840, 2160) 
    ]

    supported_resolutions = []

    for resolution in common_resolutions:
        width, height = resolution
        cap.set(cv2.CAP_PROP_FRAME_WIDTH, width)
        cap.set(cv2.CAP_PROP_FRAME_HEIGHT, height)
        actual_width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
        actual_height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
        if int(actual_width) == width and int(actual_height) == height:
            supported_resolutions.append((int(actual_width), int(actual_height)))

    print("Supported resolutions:")
    for resolution in supported_resolutions:
        print(f"{resolution[0]}x{resolution[1]}")

cap.release()
```

```python
def list_supported_capture_properties(cap: cv2.VideoCapture):
    """ List the properties supported by the capture device.
    """
    supported = list()
    for attr in dir(cv2):
        if attr.startswith('CAP_PROP'):
            if cap.get(getattr(cv2, attr)) != -1:
                supported.append(attr)
    return supported
```
